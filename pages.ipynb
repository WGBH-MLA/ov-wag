```python

def create_collection_page(collection):
    content=[
            ('text', markdownify(collection.background)) if collection.background else None,
            ('text', markdownify(collection.featured)) if collection.featured else None,
            ('text', markdownify(collection.funders)) if collection.funders else None,
            ('text', markdownify(collection.help)) if collection.help else None,
            ('text', markdownify(collection.terms)) if collection.terms else None,
            ('text', markdownify(collection.timeline)) if collection.timeline else None,
            ('text', markdownify(collection.sort)) if collection.sort else None,
            ('text', markdownify(collection.thumbnail)) if collection.thumbnail else None,
            ('text', markdownify(collection.resources)) if collection.resources else None,
        ]
    content = [item for item in content if item is not None]
    # return content
    page = OpenVaultCollection(
        title=collection.title,
        introduction=markdownify(collection.summary),
        content=content
    )
    return page

# create_collection_page(collections[0])
```

```python

for collection in collections:
    page = create_collection_page(collection)
    print(f"Creating collection: {page}")
    aapb.add_child(instance=page)
    print(f"Added collection: {collection.title}")
```
